```python
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt  
kashti = sns.load_dataset("titanic")
kashti
sns.barplot(x="sex",y="alone",hue="who" ,data=kashti, order=["feamle","male"], estimator= mean  )
plt.show()
```

```python
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt  
kashti = sns.load_dataset("titanic")
kashti
sns.barplot(x="sex",y="alone",hue="who" ,data=kashti, order=["feamle","male"], estimator= median  )
plt.show()
```
