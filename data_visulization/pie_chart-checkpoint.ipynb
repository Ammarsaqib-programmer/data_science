```python
import matplotlib.pyplot as plt

values = [20, 20, 35, 35]
plt.pie(values)
```

```python
import matplotlib.pyplot as plt

values = [20, 20, 35, 35]
plt.pie(values, labels=["A", "B", "C", "D"])
```

```python
import matplotlib.pyplot as plt

values = [20, 20, 35, 35]
plt.pie(values, labels=["A", "B", "C", "D"], explode=(0, 0.2, 0.1, 0))
```
